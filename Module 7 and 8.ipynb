```python
#str=input("Enter the security issue")
import nltk
import re
import csv
import os
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer 
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
wl=WordNetLemmatizer()

	#porter stemmer
ps=PorterStemmer()

	#list of stopwords 
stop_words = stopwords.words('english')

def clean_lemmatize(str1):
	tx=str1
		#print(tx)
	txt=re.sub(r'[^\w\s]','',tx) # to remove punctuation and replace it with space

		#print(txt)
	l=word_tokenize(txt) # splitting into words
		#print(l)
	f=[]

		#remove stop word
	for i in l:
		if i not in stop_words:
			f.append(i)
	l=[]
		


		#lemmatize the words
	for i in f:
		v=wl.lemmatize(i)
		l.append(v.lower())
	return str(l)

def clean_stemming(str1):
	tx=str1
		#print(tx)
	txt=re.sub(r'[^\w\s]',' ',tx) # to remove punctuation
		#print(txt)
	l=word_tokenize(txt) # splitting into words
		#print(l)
	f=[]

		#remove stop word
	for i in l:
		if i not in stop_words:
			f.append(i)
	l=[]
	s=''

		#lemmatize the words
	for i in f:
		v=ps.stem(i)
		l.append(v.lower())
	return str(l)


```

```python
str1=input("Enter the security issue: ")
lem=clean_lemmatize(str1)
stm=clean_stemming(str1)




file1=open("security_issues.csv", "a", encoding='utf-8')
csv_file1=csv.writer(file1)
row=[str1,'','','','',lem,stm]
csv_file1.writerow(row)
#df["Issue"][len(df)]=str1
#print(df.iloc[len(df)-1])
#import gensim
file1.close()
df=pd.read_csv("security_issues.csv")
#len(df)
```

```python
def stringlemmatize(row):
    txt=row['Lemmatize'][1:len(row['Lemmatize'])-1].split(',')

            #print(txt)
    for i in range(len(txt)):
        ind1=txt[i].find('\'')
        ind2=txt[i].rfind('\'')
        
        txt[i]=txt[i][ind1+1:ind2]
    str1=''
    for i in txt:
        str1=str1+" "+i
     
    return str1
        
        
def stringstemming(row):
    txt=row['Stemming'][1:len(row['Stemming'])-1].split(',')

            #print(txt)
    for i in range(len(txt)):
        ind1=txt[i].find('\'')
        ind2=txt[i].rfind('\'')
        txt[i]=txt[i][ind1+1:ind2]
    str1=''
    for i in txt:
        str1=str1+" "+i
     
    return str1
```

```python
df['Lemmatize']=df.apply(stringlemmatize,axis=1)
df['Stemming']=df.apply(stringstemming,axis=1)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
count=CountVectorizer()
Count_matrix=count.fit_transform(df['Lemmatize'])
cosine_matrix=cosine_similarity(Count_matrix)
cosine_matrix
```

```python
Issues_similar=list(enumerate(cosine_matrix[len(df)-1]))
Issues_similar
```

```python
sorted_issues_similar=sorted(Issues_similar,key=lambda x:x[1],reverse=True)
sorted_issues_similar=sorted_issues_similar[1:]
sorted_issues_similar
```

```python
def know_issue_from_index(index):
    return df[df.index == index]["Issue"].values[0]


print("Top 10 issues similiar are:\n")
i=0
for issues in sorted_issues_similar:
    i=i+1
    print(i,'.',know_issue_from_index(issues[0]))
    
    if i>=10:
        break
```
